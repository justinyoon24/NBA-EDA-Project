```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
# set up notebook to display multiple outputs in one cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv('nba_final.csv')
df.info()
df.head()
```

Seaborn Pair Plot Findings
1. All the variables in the pair plots are relatively normal shaped
2. There seems to be a high positive correlation between points and 3 pointers made. This makes sense since the more 3 pointers that are made, the more points that will be scored. There is also a positive correlation between assists and 3 pointers made/points. This makes sence because the more assists mean more baskets are being made. There is no real pattern between 3 pointers made and win percentage where it was expected that the more 3 pointers made would be represented by more wins. 

```python
#pair plots
sns.pairplot(df[['WIN%','3PM', 'PTS','FTM', 'REB', 'AST', 'TOV', 'STL','BLK']])
```

```python
#set up files for correlations
columns = ['WIN%','3PM','PTS','FTM', 'REB', 'AST', 'TOV', 'STL','BLK']
df_corr = df[columns]
corrmat = df_corr.corr()
mask = np.triu(np.ones_like(corrmat, dtype = bool))
f, ax = plt.subplots(figsize=(9,6))
sns.heatmap(corrmat, vmax=.8, square=True, mask = mask, annot=True, cmap='seismic', linewidths=.5)
```

Correlation Results

It makes sense that there is a high positive correlation between points and 3 pointers made because the more 3 pointers that are being made means the more points being scored. It also makes sense that there is a positive correlation between assists and points/3 pointers made since more assists likely mean better/more efficient shots are being taken. 

It suprising that there wasn't much correlation between 3 pointers made and win percentage. It was expected with the new age of basketball and analytics that the 3 pointer is one of the most efficient shots in the game, and in turn have a high positive correlation with win percentage.

There doesn't seem to be a reasonable explanation for the low negative correlation between steals and rebounds. However, the low negative correlation between win percentage and turnovers suggests that the more turnovers a team has the lower their win percentage is. 

```python
#create the figure with the subplots
f, axes = plt.subplots(1,4)
plt.figsize=(10,20)

sns.boxplot(y = df['WIN%'], orient = 'v', color = 'r', ax = axes[0])
sns.boxplot(y = df['3PM'], orient = 'v', color = 'b', ax = axes[1])
sns.boxplot(y = df['PTS'], orient = 'v', color = 'g', ax = axes[2])
sns.boxplot(y = df['FTM'], orient = 'v', color = 'y', ax = axes[3])


plt.tight_layout()
```

```python
f, axes = plt.subplots(1,5)
plt.figsize=(10,20)

sns.boxplot(y = df['REB'], orient = 'v', color = 'b', ax = axes[0])
sns.boxplot(y = df['AST'], orient = 'v', color = 'orange', ax = axes[1])
sns.boxplot(y = df['TOV'], orient = 'v', color = 'c', ax = axes[2])
sns.boxplot(y = df['STL'], orient = 'v', color = 'brown', ax = axes[3])
sns.boxplot(y = df['BLK'], orient = 'v', color = 'm', ax = axes[4])

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(9,7))
ax= sns.scatterplot(x='3PM', y='WIN%', data=df, s=50)
ax.set_xlabel('3 Pointers Made Per Game',fontsize=14)
ax.set_ylabel('Team Win Percentage', fontsize=14)
ax.set_title('3 Pointers Made Per Game and Team Win Percentage from 2011-2021 Regular Season', fontsize=15)
```

There doesn't appear to be any significant outliers in the data. The few outliers in the boxplots are considered accurate from the NBA official website.
